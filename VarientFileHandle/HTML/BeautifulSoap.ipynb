```python
import os
from bs4 import BeautifulSoup
```

```python
if os.path.exists(HTML_URL):
    with open(HTML_URL, 'r', encoding='utf-8') as file:
        content = file.read()

    soup = BeautifulSoup(content, 'html.parser')
```

```python
scripts = soup.find_all('script')
script_libs = []
script_contents = ""
for script in scripts:
    if script.get('src')!= None:
        script_libs.append(script.get('src'))
    else:
        script_contents += script.string if script.string else ""
```

```python
os.makedirs("w3resources", exist_ok=True)
with open("w3resources/inline_script.js", "w", encoding='utf-8') as f:
    f.write("\n".join(script_libs))
    f.write(script_contents)
```

```python
styles= soup.find_all('style')
sytles_libs = []
style_contents = ""
for style in styles:
    if style.get('href')!= None:
        sytles_libs.append(style.get('href'))
    else:
        style_contents += style.string if style.string else ""
```

```python
with open("w3resources/inline_styles.css", "w", encoding='utf-8') as f:
    f.write("\n".join(sytles_libs))
    style_contents.replace('!important;', '')
    f.write(style_contents)
```

```python
meta_details = soup.find_all('meta')
title = soup.title.string if soup.title else "No Title Found"
meta_infos = {"title":title}
for meta in meta_details:
    if  meta.get('content')!= None:
        key=meta.get('name') or meta.get('property')
        meta_infos[key] =  meta.get('content')
```

```python
icon= soup.find('link', rel='shortcut icon')
meta_infos['favicon'] = icon.get('href') if icon else None
```

```python
import json
```

```python
with open("w3resources/meta_info.json", "w", encoding='utf-8') as f:
    json.dump(meta_infos, f, indent=4)
```

```python
Body = soup.find('body')
# Remove all <script> and <style> tags inside the body
if Body is not None:
    for tag in Body.find_all(['script', 'style']):
        tag.decompose()
```
