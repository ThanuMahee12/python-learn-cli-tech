```python
import pandas as pd
```

```python
response = requests.get('https://jsonplaceholder.typicode.com/todos')
print("statuscode",response.status_code,sep=":")
print("response",response.json(),sep=":")
print("header",response.headers,sep=":")
print("url",response.url,sep=":")
print("encoding",response.encoding,sep=":")
print("content",response.content,sep=":")
print("text",response.text,sep=":")
print("cookies",response.cookies.get_dict(),sep=":")
print("history",response.history,sep=":")
print("elapsed",response.elapsed,sep=":")
print("reason",response.reason,sep=":")
print("request",response.request.url,sep=":")
```

```python
post_response = requests.get('https://jsonplaceholder.typicode.com/posts')
post_response_data = post_response.json()
df=pd.DataFrame(post_response_data)
df.to_csv('../asserts/posts.csv',index=False)
```

```python
post_data_response = requests.post('https://jsonplaceholder.typicode.com/posts', data={'title': 'foo', 'body': 'bar', 'userId': 1}) 
print("statuscode",post_data_response.status_code,sep=":")
print("response",post_data_response.json(),sep=":")
```
